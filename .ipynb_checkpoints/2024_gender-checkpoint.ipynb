```python
import csv
import matplotlib.pyplot as plt

f = open("datas/202403_gender.csv", 'r')
data = csv.reader(f)

#print(data)
#for row in data:
#    print(row[0])

m = []
f = []
for row in data:
    if "" in row[0]:
        for i in range(0, 101):
            m.append(int(row[i+3]))  # 0  
            f.append(int(row[-(i+1)]))
f.reverse()  # f  ()
print(len(m), len(f))
print(m)
print(f)

plt.rc('font', family='Malgun Gothic')
plt.title("     ")
plt.plot(m, label="")
plt.plot(f, label="")
plt.xlabel("")
plt.ylabel("")
# plt.bar(range(101), m, label="")  #   (barh - )
# plt.bar(range(101), f, label="")
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

f = open("datas/202403_gender.csv", 'r')
data = csv.reader(f)

#print(data)
#for row in data:
#    print(row[0])

m = []
f = []
name = input(" : ")
for row in data:
    if name in row[0]:
        for i in range(0, 101):
            m.append(int(row[i+3].replace(',', '')))  # 0  (  )
            f.append(int(row[-(i+1)].replace(',', '')))
        break
f.reverse()  # f  ()
print(len(m), len(f))
print(m)
print(f)

plt.rc('font', family='Malgun Gothic')
plt.title(name + "     ")
plt.plot(m, label="")
plt.plot(f, label="")
plt.xlabel("")
plt.ylabel("")
# plt.bar(range(101), m, label="")  #   (barh - )
# plt.bar(range(101), f, label="")
plt.show()
```
