```python
import csv
import matplotlib.pyplot as plt

f = open("datas/202403_gender.csv", 'r')
data = csv.reader(f)

#print(data)
#for row in data:
#    print(row[0])

m = []
f = []
for row in data:
    if "신도림" in row[0]:
        for i in range(0, 101):
            m.append(int(row[i+3]))  # 0세부터 정수로 변환
            f.append(int(row[-(i+1)]))
f.reverse()  # f 리스트 거꾸로(뒤집기)
print(len(m), len(f))
print(m)
print(f)

plt.rc('font', family='Malgun Gothic')
plt.title("신도림 지역의 남녀 성별 인구 분포")
plt.plot(m, label="남성")
plt.plot(f, label="여성")
plt.xlabel("나이")
plt.ylabel("인구수")
# plt.bar(range(101), m, label="남성")  # 세로 막대 그래프(barh - 가로막대)
# plt.bar(range(101), f, label="여성")
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

f = open("datas/202403_gender.csv", 'r')
data = csv.reader(f)

#print(data)
#for row in data:
#    print(row[0])

m = []
f = []
name = input("지역이름 입력: ")
for row in data:
    if name in row[0]:
        for i in range(0, 101):
            m.append(int(row[i+3].replace(',', '')))  # 0세부터 정수로 변환(천단위 쉼표 제거)
            f.append(int(row[-(i+1)].replace(',', '')))
        break
f.reverse()  # f 리스트 거꾸로(뒤집기)
print(len(m), len(f))
print(m)
print(f)

plt.rc('font', family='Malgun Gothic')
plt.title(name + " 지역의 남녀 성별 인구 분포")
plt.plot(m, label="남성")
plt.plot(f, label="여성")
plt.xlabel("나이")
plt.ylabel("인구수")
# plt.bar(range(101), m, label="남성")  # 세로 막대 그래프(barh - 가로막대)
# plt.bar(range(101), f, label="여성")
plt.show()
```
