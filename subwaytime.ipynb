### 출근 시간대 사람들이 가장 많이 타고 내리는 역은 어디일까?

```python
import csv
import matplotlib.pyplot as plt

f = open("datas/subwaytime.csv", "r")
data = csv.reader(f)
next(data)
next(data)
result = []  # 승차 데이터 저장
for row in data:
    for i in range(48):
        row[i+4] = row[i+4].replace(',', '') # 천단위 구분기호(,) 제거
    row[4:] = map(int, row[4:])  # 문자를 정수로 변환
    result.append(row[10])  # 아침 7시 승차 데이터
print(len(result)) # 역 개수
#print(result)

# 막대그래프로 보기
result.sort()  #오름차순 정렬
plt.bar(range(len(result)), result)
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

f = open("datas/subwaytime.csv", "r")
data = csv.reader(f)
next(data)
next(data)
result = []  # 승차 데이터 저장

for row in data:
    for i in range(48):
        row[i+4] = row[i+4].replace(',', '') # 천단위 구분기호(,) 제거
    row[4:] = map(int, row[4:])  # 문자를 정수로 변환
    result.append(sum(row[10:15:2]))  # 아침 7시~9시 승차인원 합계 데이터(10번, 12번, 14번)

# 막대그래프로 보기
result.sort()  #오름차순 정렬
plt.bar(range(len(result)), result)
plt.show()
```
