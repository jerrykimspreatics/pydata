### 최고 기온을 꺽은선 그래프로 시각화하기

```python
import csv
import matplotlib.pyplot as plt

with open("datas/seoul_2024.csv", "r") as f:
    data = csv.reader(f)
    next(data)
    
    temps = []   # 최고 기온 저장

    for row in data:
        if row[-1] != "":  # 최고 기온이 존재한다면
            #print(float(row[-1]))  # 실수로 출력
            temps.append(float(row[-1]))
    print(len(temps))
    print(temps)
    plt.plot(temps, 'r')
    plt.show()
```

### 내 생일의 최고기온을 시각화하기

```python
# 날짜 데이터 추출하기
'''
date = "1907-10-01"
print(date.split('-'))
print(date.split('-')[0])
print(date.split('-')[1])
'''
import csv
import matplotlib.pyplot as plt

with open("datas/seoul_2024.csv", "r") as f:
    data = csv.reader(f)
    next(data)
    high = []   # 최고 기온 
    low = []    # 최저 기온

    for row in data:
        if row[-1] != "" and row[-2] != "":  # 최고 기온과 최저 기온이 존재한다면
            if 1970 <= int(row[0].split('-')[0]):  # 1970년 이후 데이터라면
                if row[0].split('-')[1] == '02' and row[0].split('-')[2] == '14':
                    high.append(float(row[-1]))
                    low.append(float(row[-2]))

    plt.rc('font', family='Malgun Gothic')      #한글 글꼴 추가
    plt.rcParams['axes.unicode_minus'] = False  #마이너스 기호 깨짐 방지
    plt.title("내 생일의 기온 변화 그래프")
    plt.plot(high, 'red', label='high')
    plt.plot(low, 'blue', label='low')
    plt.xlabel("연도")
    plt.ylabel("기온")
    plt.legend()
    plt.show()
```
