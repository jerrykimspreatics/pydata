```python
import csv
import matplotlib.pyplot as plt

f = open("datas/age_2023.csv", "r")
data = csv.reader(f)
ages = []

for row in data:
    if "신도림" in row[0]:
        for i in row[3:]:  # 0~100세 
            #print(row)
            ages.append(int(i))

plt.style.use('ggplot')
plt.plot(ages)
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

f = open("datas/age_2023.csv", "r")
data = csv.reader(f)
ages = []
name = input('지역 이름: ')

for row in data:
    if name in row[0]:
        for i in row[3:]:  # 0~100세 
            ages.append(int(i))

#print(ages)
plt.style.use('ggplot')
plt.plot(ages)
plt.show()
```
