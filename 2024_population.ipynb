```python
import csv

with open('datas/202406_population.csv', 'r') as f:
    data = csv.reader(f)
    header = next(data)
    # print(data)
    number = [] #연령별 인구수를 담을 배열 선언
    
    for row in data:
        if '신도림' in row[0]:
            for i in row[3:]: #0 ~ 100세
                # print(i)
                number.append(int(i))  # 문자를 정수로 변환
                
print(number)
```

### 연령별 인구분포 그래프

```python
import csv
import matplotlib.pyplot as plt

with open('datas/202406_population.csv', 'r') as f:
    data = csv.reader(f)
    header = next(data)
    # print(data)
    number = [] #연령별 인구수를 담을 배열 선언
    name = input("지역의 이름(읍면동 단위)을 입력하세요: ")
    
    for row in data:
        if '신도림' in row[0]:
            for i in row[3:]: #0 ~ 100세
                # print(i)
                number.append(int(i))
# print(number)
plt.style.use('ggplot')  #격자 무늬
plt.rc('font', family='Malgun Gothic') #한글 글꼴 설정
plt.title(name + " 지역의 인구 구조")
plt.plot(number)
plt.show()
```
