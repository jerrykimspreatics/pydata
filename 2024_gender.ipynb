### 성별 인구를 파이차트로 그리기

```python
import csv
import matplotlib.pyplot as plt

with open("datas/202403_gender.csv", 'r') as f:
    data = csv.reader(f)

    #print(data)
    #for row in data:
    #    print(row[0])
    
    male = 0     # 남성 인구수
    female = 0   # 여성 인구수
    size = []    # 냠녀 인구수를 저장할 리스트
    name = input("지역이름 입력: ")
    for row in data:
        if name in row[0]:
            for i in range(0, 101):
               male += int(row[i+3].replace(',', '')) # 0세부터 정수로 변환(천단위 쉼표 제거)
               female += int(row[-(i+1)].replace(',', ''))
            break

    print(male)
    print(female)
    size.append(male)
    size.append(female)

# 파이 차트
plt.rc('font', family="Malgun Gothic")
color = ['crimson', 'darkcyan']
plt.axis('equal')
plt.title(name + " 지역의 남녀 성별 비율")
plt.pie(size, autopct="%.1f%%", labels=['남', '여'], startangle=90 )
plt.show()
```

### 제주도의 연령대별 성별 비율을 산점도로 표현

```python
import csv
import matplotlib.pyplot as plt
import math

with open('datas/202403_gender.csv', 'r') as f:
    data = csv.reader(f)
    m = []
    f = []
    size = []
    name = input("지역 이름 입력: ")
    for row in data:
        if name in row[0]:
            for i in range(3, 104):
                m.append(int(row[i]))
                f.append(int(row[i+103]))
                size.append(math.sqrt(int(row[i]) + int(row[i+103])))
            break
plt.scatter(m, f, s=size, c=range(101), alpha=0.5, cmap='jet')
plt.colorbar()
plt.plot(range(max(m)), range(max(m)), 'g')
plt.show()
```
