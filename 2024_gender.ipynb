```python
import csv
import matplotlib.pyplot as plt

with open("datas/202403_gender.csv", 'r') as f:
    data = csv.reader(f)

    #print(data)
    #for row in data:
    #    print(row[0])
    
    male = 0     #  
    female = 0   #  
    size = []    #    
    name = input(" : ")
    for row in data:
        if name in row[0]:
            for i in range(0, 101):
               male += int(row[i+3].replace(',', '')) # 0  (  )
               female += int(row[-(i+1)].replace(',', ''))
            break

    print(male)
    print(female)
    size.append(male)
    size.append(female)

#  
plt.rc('font', family="Malgun Gothic")
color = ['crimson', 'darkcyan']
plt.axis('equal')
plt.title(name + "    ")
plt.pie(size, autopct="%.1f%%", labels=['', ''], startangle=90 )
plt.show()
```

```python
import matplotlib.pyplot as plt
import random

x = []
y = []
size = []
for i in range(100):
    x.append(random.randint(50, 100))
    y.append(random.randint(50, 100))
    size.append(random.randint(10, 100))

plt.scatter(x, y, s=size, c=size, cmap='jet', alpha=0.6)
plt.colorbar()
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt
import math

with open('datas/202403_gender.csv', 'r') as f:
    data = csv.reader(f)
    m = []
    f = []
    size = []
    name = input("  : ")
    for row in data:
        if name in row[0]:
            for i in range(3, 104):
                m.append(int(row[i]))
                f.append(int(row[i+103]))
                size.append(math.sqrt(int(row[i]) + int(row[i+103])))
            break
plt.scatter(m, f, s=size, c=range(101), alpha=0.5, cmap='jet')
plt.colorbar()
plt.plot(range(max(m)), range(max(m)), 'g')
plt.show()
```
